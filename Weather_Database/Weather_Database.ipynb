```python
# Use the citipy module to determine city based on latitude and longitude.
from citipy import citipy
# Import the random module.
import random
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
# Dependencies
import requests
import json #stores info in dictionaries
```

```python
# Create a set of random latitude and longitude combinations.
lats = np.random.uniform(low=-90.000, high=90.000, size=2000)
lngs = np.random.uniform(low=-180.000, high=180.000, size=2000)
lat_lngs = zip(lats, lngs)
```

```python
# Create a list for holding the cities.
cities = []
for coordinate in lat_lngs:
    city = citipy.nearest_city(coordinate[0], coordinate[1]).city_name
    # If the city is unique, then we will add it to the cities list.
    if city not in cities:
        cities.append(city)
```

```python
# Import the requests library.
import requests
# Import the API key.
from config import weather_api_key
```

```python
# Starting URL for Weather Map API Call.
base_url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + weather_api_key
```

```python
print("Beginning Data Retrieval     ")
print("------------------------------")
city_data = []
for city in cities:
    city_info = base_url + "&q=" + city
    geo_data = requests.get(city_info).json()
    try:
        #City description elements
        city_name = geo_data["name"]
        country = geo_data["sys"]["country"]
        longitude = geo_data["coord"]["lon"]
        latitude = geo_data["coord"]["lat"]
        temp_max = geo_data["main"]["temp_max"]
        humidity = geo_data["main"]["humidity"]
        cloudiness = geo_data["clouds"]["all"]
        wind_speed = geo_data["wind"]["speed"]
        current_descr = geo_data["weather"][0]["description"]
        
        
        #Creation of dictionary to append values
        city_data.append({"City":city_name,
                          "Country":country,
                          "Longitude":longitude,
                          "Latitude":latitude,
                          "Max Temp":temp_max,
                          "Humidity":humidity,
                          "Cloudiness":cloudiness,
                          "Wind Speed":wind_speed,
                          "Current Description":current_descr})
        
    except:
        print("City not found. Skipping")
        pass
print("--------------------------------")
print("Data Retrieval Complete         ")
print("---------------------------------")
```

```python
weather_db_df = pd.DataFrame(city_data)
weather_db_df.head(10)
```

```python
# Create the output file (CSV).
output_data_file = "WeatherPy_Database.csv"
# Export the City_Data into a CSV.
weather_db_df.to_csv(output_data_file, index_label="City_ID")
```
